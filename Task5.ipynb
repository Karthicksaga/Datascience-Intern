```python
#importing the required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("SampleSuperstore.csv")
df.head()
```

```python
df.info()
```

```python
print("Dimension of a Dataset:",df.shape)
print('Statistical measures of a Dataset:',df.describe())
```

```python
df.columns=df.columns.str.replace(' ','_')
df.head()
```

```python
#check if there is any missing values in the dataset
print(df.isnull().sum())
```

```python
print(df.info())
```

```python
#Numerical features in the dataset
numerical_features=[feature for feature in df.columns if df[feature].dtypes!='O']
print("Numerical Features:",numerical_features)
```

```python
discrete_features=[feature for feature in df.columns if len(df[feature].unique())<30]
print("Discretefeatures",discrete_features)
```

```python
for feature in discrete_features:
    if feature not in "Profit":
        plt.figure(figsize=(10,10))
        plt.title("Bar plot of %s"%(feature))
        sns.barplot(x=df[feature],y=df['Profit'])
        plt.xlabel(feature)
       
```

Our company should mainly focus on the furniture,officesupplies category and the east region to increase the profit compared to other features

```python
for feature in discrete_features:
    if feature not in "Profit":
        plt.figure(figsize=(10,10))   
        plt.title("PieChart")
        df[feature].value_counts().plot.pie(autopct="%4.1f%%",startangle=90,shadow=True)
        
```

```python
for feature in discrete_features:
    if feature not in "Profit":
        plt.figure(figsize=(10,10))
        plt.title("Bar plot of %s"%(feature))
        sns.barplot(x=df[feature],y=df['Sales'])
        plt.xlabel(feature)
       
```

```python
continous_features=[feature for feature in df.columns if feature not in discrete_features]
print("Continous Features:",continous_features)
```

```python
for feature in continous_features:
    if feature not in "Profit":
        plt.figure(figsize=(15,10))
        plt.subplot(1,2,1)
        plt.title("Distribution of %s"%(feature))
        sns.scatterplot(x=df[feature],y=df["Profit"])
       
```

```python
for feature in continous_features:
    if feature not in "Sales":
        plt.figure(figsize=(15,10))
        plt.title("Distribution of %s"%(feature))
        sns.scatterplot(x=df[feature],y= df["Sales"])
       
```

```python
plt.figure(figsize = (10,10))
sns.distplot(df.Profit ,hist = False )
sns.distplot(df.Sales ,hist =  False)
plt.title("Profit & Sales")
plt.show()
```

```python
plt.figure(figsize = (10,10))
top = df.groupby("State")["Profit"].sum().sort_values(ascending = False).head(10).reset_index()
#top.head()
sns.barplot(x = "State" , y = "Profit" ,data = top)
plt.title("Profit Earned by  Top 10 State")
plt.xlabel("State")
plt.ylabel("Profit")
```

```python
plt.figure(figsize = (20,10))
top10_city = df.groupby("City")["Profit"].sum().sort_values(ascending = False).head(10).reset_index()
#top.head()
sns.barplot(x = "City" , y = "Profit" ,data = top10_city)
plt.title("Profit Earned by  Top 10 city")
plt.xlabel("State")
plt.ylabel("Profit")
```

```python
columns = ["City","Category"]
top10_city = df.groupby(columns)["Sales"].sum().sort_values(ascending = False).head(10).reset_index()
top10_city
```

```python
plt.figure(figsize = (20,10))
columns = ["City","Category"]
top10_city_category = df.groupby(columns)["Sales"].sum().sort_values(ascending = False).head(10).reset_index()
#top.head()
sns.barplot(x = "City" , y = "Sales",hue = "Category" ,data = top10_city_category)
plt.title("Sales by Top City based on Category")
plt.xlabel("City")
plt.ylabel("Sales")
```

```python
top10_city_category_profit = df.groupby(columns)["Profit"].sum().sort_values(ascending = False).head(10).reset_index()
top10_city_category_profit
```

```python
plt.figure(figsize = (20,10))
columns = ["City","Category"]
top10_city_category_profit = df.groupby(columns)["Profit"].sum().sort_values(ascending = False).head(10).reset_index()
#top.head()
sns.barplot(x = "City" , y = "Profit",hue = "Category" ,data = top10_city_category_profit)
plt.title("Sales by Top City based on Category")
plt.xlabel("City")
plt.ylabel("Profit")
```
